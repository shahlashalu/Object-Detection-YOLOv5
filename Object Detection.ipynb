```python
# import necessary libraries
import torch
from pathlib import Path
import cv2
from matplotlib import pyplot as plt
```

```python
# load YOLOv5 pretrained model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained = True)
```

```python
# load an image for Object Detection
image_path = r"C:\Users\hp\objectdetection.jpg" # Read the image
image = cv2.imread(image_path)

# Perform object detection
results = model(image_path)   # Detect Objects

# Display results
results.show()   # Show detection on the image
results.print()  # print detection details
```
